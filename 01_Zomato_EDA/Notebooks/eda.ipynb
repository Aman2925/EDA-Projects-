# Final Dataset

```python
final_data = pd.merge(left=data,right=data_2,on='Country Code',how='left')
```

```python
final_data.head()
```

```python
# Checking data types 

final_data.dtypes
```

```python
country_names = final_data.Country.value_counts().index[:3]
```

```python
country_vals = final_data.Country.value_counts().values[:3]
```

```python
# Pie Chart

plt.pie(x=country_vals,labels=country_names,autopct='%1.2f%%')
plt.plot
```

```python
final_data
```

```python
ratings = final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
sns.barplot(
    x='Aggregate rating',
    y='Rating Count',
    data=ratings,
    hue='Rating color',
    palette=['#f5f5f5', '#d62728', '#ff7f0e', '#ffdd57', '#2ca02c', '#006400']
)
plt.show()
```

```python
sns.countplot(
    x='Rating color',
    data=ratings,
    palette=['#f5f5f5', '#d62728', '#ff7f0e', '#ffdd57', '#2ca02c', '#006400']
)
plt.show()
```

## Countries with zero ratings

```python
final_data.Country.value_counts()
```

```python
zero_rating_countries = final_data[final_data['Aggregate rating'] == 0] \
    .groupby('Country', as_index=False) \
    .size() \
    .rename(columns={'size': 'Zero Rating Records'})

zero_rating_countries
```
